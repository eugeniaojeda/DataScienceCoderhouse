# ***Análisis Exploratorio de Datos (EDA)***


A continuación, revisamos la distribución de los casos confirmados de COVID-19 en Argentina:

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(x="fecha_apertura", data=df)
plt.title("Distribución de casos confirmados de COVID-19 en Argentina")
plt.xticks(rotation=90)
plt.show()
```

Podemos ver que el número de casos confirmados de COVID-19 ha aumentado rápidamente desde que se reportó el primer caso en marzo de 2020, y ha tenido varios picos durante la pandemia.

Este gráfico muestra la evolución del número de casos diarios.

```python
casos_por_dia = df.groupby('fecha_apertura').size()
plt.plot(casos_por_dia.index, casos_por_dia.values)
plt.title('Evolución del número de casos diarios')
plt.xlabel('Fecha')
plt.ylabel('Número de casos')
plt.xticks(rotation=90)
plt.show()
```

A continuación, revisamos la tasa de mortalidad de COVID-19 en Argentina:
Podemos ver que la tasa de mortalidad de COVID-19 en Argentina es del 1.33%.

```python
mortalidad = df[df['fallecido'] == 'SI'].shape[0]/df.shape[0]*100
print(f"La tasa de mortalidad de COVID-19 en Argentina es del {round(mortalidad, 2)}%")
```

A continuación, revisamos la distribución de los casos confirmados de COVID-19 por provincia:

```python
plt.figure(figsize=(12,6))
sns.countplot(x="provincia_residencia", data=df, order=df['provincia_residencia'].value_counts().index)
plt.title("Distribución de casos confirmados de COVID-19 por provincia")
plt.xticks(rotation=90)
plt.show()
```

A continuación, revisamos la distribución de los casos confirmados de COVID-19 por edad y sexo:

```python
plt.figure(figsize=(12,6))
sns.histplot(x="edad", data=df, hue="sexo", kde=True, bins=20)
plt.title("Distribución de casos confirmados de COVID-19 por edad y sexo")
plt.show()
```

```python
sexos = df['sexo'].value_counts()
plt.bar(sexos.index, sexos.values)
plt.title('Número de casos por sexo')
plt.xlabel('Sexo')
plt.ylabel('Número de casos')
plt.show()
```

Este gráfico muestra la proporción de casos por grupo etario.

```python
# Crear grupos etarios
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '61+']
df['grupo_etario'] = pd.cut(df['edad'], bins=bins, labels=labels)

# Calcular el número de casos por grupo etario
edades = df['grupo_etario'].value_counts()

# Crear el gráfico de pastel
plt.pie(edades.values, labels=edades.index, autopct='%1.1f%%')
plt.title('Proporción de casos por grupo etario')
plt.show()
```

Aquí se utiliza un gráfico de dispersión tridimensional, usando la edad, la provincia de residencia y la provincia de carga como nuestras tres variables de interés.

```python
# crear figura y ejes 3D
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# asignar variables a ejes
x = df['provincia_residencia_codigo_indec']
y = df['provincia_carga_codigo_indec']
z = df['edad']

# crear gráfico de dispersión tridimensional
ax.scatter(x, y, z, s=10, alpha=0.5)

# agregar etiquetas de eje y título
ax.set_xlabel('Provincia de residencia')
ax.set_ylabel('Provincia de carga')
ax.set_zlabel('Edad')
ax.set_title('Gráfico de dispersión tridimensional')

# mostrar gráfico
plt.show()
```

Aquí se realiza una visualización con respecto a los casos de personas fallecidas

```python
plt.figure(figsize=(8,5))
sns.barplot(x='fallecido',y='edad',data=df, palette='rainbow', hue='sexo')
plt.title("Fallecidos por edad, segmentado por sexo")
```

gráfico de barras apiladas para comparar la cantidad de casos confirmados, recuperados y fallecidos por provincia de residencia.

Para esto, vamos a utilizar las columnas 'provincia_residencia', 'clasificacion_resumen' y 'numero_de_caso' del dataset. Este gráfico nos permite comparar la cantidad de casos confirmados, recuperados y fallecidos por provincia de residencia de manera clara y visual.

```python
# Filtrar solo casos confirmados, recuperados y fallecidos
df = df[df['clasificacion_resumen'].isin(['Confirmado', 'Recuperado', 'Fallecido'])]

# Agrupar los datos por provincia y clasificación resumen, y sumar la cantidad de casos
grouped_data = df.groupby(['provincia_residencia', 'clasificacion_resumen']).agg({'numero_de_caso': 'count'})

# Crear el gráfico de barras apiladas
grouped_data.unstack().plot(kind='bar', stacked=True)

# Configurar el gráfico
plt.title('Casos confirmados, recuperados y fallecidos por provincia de residencia')
plt.xlabel('Provincia de residencia')
plt.ylabel('Cantidad de casos')
plt.legend(['Confirmado', 'Recuperado', 'Fallecido'])
plt.show()
```

```python
import pandas as pd
import requests
import json

# Obtener la API mediante registro en el sitio web
api_key = '4515ff01dfde41be977f107f6d04b722'

# Definir los parámetros de la solicitud de noticias
parameters = {
    'q': 'COVID-19',  # Palabra clave para buscar noticias relacionadas con COVID-19
    'language': 'es',  # Idioma de las noticias (español)
    'pageSize': 100  # Número máximo de noticias a obtener (máximo: 100)
}

# Hacer la solicitud a la API de noticias
response = requests.get(f'https://newsapi.org/v2/everything?apiKey={api_key}', params=parameters)

# Obtener los datos en formato JSON
data = json.loads(response.text)

# Extraer los datos relevantes para el nuevo dataframe
articles = data['articles']
df = pd.DataFrame(articles)

# Explorar el nuevo dataframe
print(df.head())  # Muestra las primeras filas del dataframe
print(df.shape)  # Muestra el número de filas y columnas
print(df.dtypes)  # Muestra los tipos de datos de cada columna

# Almacenar los datos en un archivo CSV
df.to_csv('news_data.csv', index=False)
```

```python
from reportlab.lib.pagesizes import letter
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, Image
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
import matplotlib.pyplot as plt
import pandas as pd
from io import BytesIO
# Cargar el archivo CSV
df = pd.read_csv('news_data.csv')

# Función para crear el gráfico de barras de la distribución de noticias por mes
def create_bar_chart():
    plt.figure(figsize=(10, 5))
    df['publishedAt'] = pd.to_datetime(df['publishedAt'])
    monthly_counts = df['publishedAt'].dt.month.value_counts().sort_index()
    plt.bar(monthly_counts.index, monthly_counts.values)
    plt.xlabel('Mes')
    plt.ylabel('Cantidad de noticias')
    plt.title('Distribución de noticias por mes')
    plt.xticks(monthly_counts.index)
    return plt

# Función para crear el análisis de sentimiento de los títulos de las noticias
def sentiment_analysis():
    from textblob import TextBlob
    df['sentiment'] = df['title'].apply(lambda x: TextBlob(x).sentiment.polarity)
    return df[['title', 'sentiment']]

# Función para generar el informe en formato PDF
def generate_report():
    doc = SimpleDocTemplate("informe_analisis_noticias.pdf", pagesize=letter)
    styles = getSampleStyleSheet()

    # Crear el contenido del informe
    story = []

    # Título del informe
    title_style = ParagraphStyle(name='TitleStyle', fontSize=20, textColor=colors.darkblue, spaceAfter=20)
    story.append(Paragraph("Informe de Análisis de Noticias sobre COVID-19", title_style))

    # Resumen Ejecutivo
    summary_style = ParagraphStyle(name='SummaryStyle', fontSize=12, spaceAfter=20)
    summary_text = "Este informe presenta un análisis de noticias relacionadas con COVID-19 obtenidas a través de la API de noticias. Se analizaron los datos para obtener información relevante sobre las noticias, como la cantidad total de noticias, las fuentes más frecuentes, la distribución temporal y un análisis de sentimiento de los títulos de las noticias."
    story.append(Paragraph(summary_text, summary_style))

    # Sección 1: Introducción
    introduction_text = "1. Introducción:\nLa pandemia de COVID-19 ha sido un evento de gran relevancia a nivel mundial, y las noticias relacionadas con esta enfermedad han sido abundantes. El objetivo de este análisis es proporcionar información sobre la cobertura de noticias en diferentes aspectos relacionados con COVID-19."
    story.append(Paragraph(introduction_text, styles['Normal']))

    # Sección 2: Metodología
    methodology_text = "2. Metodología:\nPara realizar este análisis, se utilizó la API de noticias con la palabra clave 'COVID-19' y el idioma español. Se obtuvo un total de {} noticias relevantes. Estos datos se almacenaron en un archivo CSV para su posterior análisis.".format(df.shape[0])
    story.append(Paragraph(methodology_text, styles['Normal']))

    # Sección 3: Exploración de Datos
    exploration_text = "3. Exploración de Datos:\nNúmero total de noticias: {}\nNúmero de fuentes únicas: {}\nRango de fechas de noticias: {} a {}".format(df.shape[0], len(df['source.name'].unique()), df['publishedAt'].min(), df['publishedAt'].max())
    story.append(Paragraph(exploration_text, styles['Normal']))

# Sección 4: Análisis de Fuentes de Noticias
sources_text = "4. Análisis de Fuentes de Noticias:\nSe realizó un análisis para determinar las fuentes más frecuentes de noticias relacionadas con COVID-19. Las principales fuentes identificadas fueron:"
story.append(Paragraph(sources_text, styles['Normal']))

# Crear tabla con los resultados del análisis de fuentes
source_counts = df['source.name'].value_counts()
data = [['Fuente', 'Cantidad de Noticias']]
for source, count in source_counts.iteritems():
    data.append([source, count])

table_style = TableStyle([('BACKGROUND', (0, 0), (-1, 0), colors.lightgrey),
                          ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
                          ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
                          ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
                          ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
                          ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
                          ('GRID', (0, 0), (-1, -1), 1, colors.black)])

table = Table(data, style=table_style, colWidths=[300, 100])
story.append(table)

# Sección 5: Distribución Temporal
story.append(Paragraph("5. Distribución Temporal:", styles['Normal']))

# Crear gráfico de barras con la distribución temporal
plt = create_bar_chart()
plt_io = BytesIO()
plt.savefig(plt_io, format='png')
plt_io.seek(0)
plt.close()

# Insertar el gráfico de barras en el informe como una imagen
story.append(Spacer(1, 20))
story.append(Paragraph("Gráfico de la distribución de noticias por mes:", styles['Normal']))
story.append(Spacer(1, 10))
story.append(Image(plt_io, width=500, height=300))
plt_io.close()

# Sección 6: Análisis de Sentimiento
story.append(Paragraph("6. Análisis de Sentimiento:", styles['Normal']))

# Obtener el análisis de sentimiento y crear una tabla
sentiment_df = sentiment_analysis()
data_sentiment = [['Título de Noticia', 'Polaridad']]
for index, row in sentiment_df.iterrows():
    data_sentiment.append([row['title'], row['sentiment']])

table_sentiment = Table(data_sentiment, style=table_style, colWidths=[400, 100])
story.append(table_sentiment)

# Sección 7: Conclusiones
conclusions_text = "7. Conclusiones:\nBasado en el análisis realizado, se puede concluir lo siguiente:\n- La cobertura de noticias sobre COVID-19 ha sido extensa, con un total de {} noticias analizadas.\n- {} y {} son las fuentes más frecuentes de noticias relacionadas con COVID-19.\n- Se observa una variación en la cantidad de noticias a lo largo del tiempo, con picos en determinados meses.\n- El análisis de sentimiento sugiere que la mayoría de los títulos de noticias tienen un tono neutral.".format(df.shape[0], source_counts.index[0], source_counts.index[1])

story.append(Paragraph(conclusions_text, styles['Normal']))



# Crear el documento PDF
doc.build(story)
```
